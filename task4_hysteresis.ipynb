Hysteresis loops are graphical representations of the magnetic properties of materials, specifically showing how the magnetization of a material changes in response to an applied magnetic field. These loops provide critical information about the magnetic behavior of materials, including their coercivity, remanence, and magnetic saturation. 

The process to obtain a hysteresis loop involves applying a cyclic magnetic field to a magnetic material and measuring its magnetization. As the magnetic field is varied, the magnetization of the material changes, tracing a loop on the graph. The main features and parameters that can be derived from a hysteresis loop include:

1. **Coercivity (Hc)**: This is the measure of the resistance of a magnetic material to changes in magnetization. It is defined as the intensity of the applied magnetic field required to reduce the magnetization of the material to zero after it has been magnetized to saturation. Coercivity is represented by the horizontal distance from the origin to the hysteresis loop on the H (magnetic field strength) axis.

2. **Remanence (Br or Mr)**: After the external magnetic field is removed, the material retains some magnetization. This residual magnetization is known as remanence. It is represented by the vertical distance from the origin to the hysteresis loop on the M (magnetization) axis when the applied field is zero.

3. **Saturation Magnetization (Ms)**: This is the maximum magnetization a material can achieve under an applied magnetic field. Beyond a certain field strength, further increases in the applied field do not increase the magnetization.

4. **Energy Loss**: The area within the hysteresis loop represents the energy lost in the form of heat (hysteresis loss) during one complete cycle of magnetization and demagnetization. This energy loss is crucial in applications where energy efficiency is important, such as in transformers and motors.

Hysteresis loops can vary significantly in shape depending on the material's properties:
- **Soft magnetic materials** have narrow loops, indicating low coercivity and low hysteresis losses, which makes them suitable for applications requiring frequent magnetization and demagnetization cycles, such as in the cores of electrical transformers.
- **Hard magnetic materials** exhibit wide loops, indicating high coercivity and high remanence, making them suitable for permanent magnets where a stable magnetization is required.

Understanding the hysteresis loop of a material is essential for predicting its behavior in magnetic applications, allowing for the design and optimization of magnetic devices based on the material's magnetic properties.

### Visualizing the loop
The represent the hysteresis loop, we plot the magnetization as a function of the applied field:

````python
def hysteresis(path:str):
    task = op(path) # Open the simulation data in python
    fig, ax = plt.subplots(1,1, figsize=(5, 5)) # create empty figure
    x = task.table['B_extx'][:] # extract the external field from the table
    y = task.table['mx'][:] # extract the magnetization from the table
    ax.plot(x, y) # plot the magnetization
    ax.set_xlabel('B_extx (T)') # set the x-axis label
    ax.set_ylabel('mx') # set the y-axis label
    ax.grid(True) # show the grid
    plt.tight_layout() # make the plot pretty

hysteresis("./simulations/task4.zarr") # call the function
````
